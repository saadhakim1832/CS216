```python
import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
st_assess = pd.read_csv("OULAD/studentAssessment.csv")
st_assess.shape
```

```python
st_info = pd.read_csv("OULAD/studentInfo.csv")
st_info.shape
```

```python
st_merge = pd.merge(st_assess, st_info, on = "id_student")
st_merge["Passed"] = st_merge["final_result"] == "Pass"
st_merge.head()

query_imd = re.compile(r"")
```

```python
cd_merge = st_merge.groupby("imd_band").count()["id_student"]
cd_merge
```

```python
assessments = st_merge.groupby("id_assessment").aggregate({"id_student":"count", "score":"mean"})
assessments
```

```python
#mods = pd.cut(st_merge["code_module"], ["AAA", "BBB", "CCC", "DDD", "EEE", "FFF", "GGG"])
score_table = st_merge.pivot_table("score", index=["gender", "code_module"], columns="age_band")
score_table
```

```python
pass_table = st_merge.pivot_table("Passed", index=["gender", "code_module"], columns="age_band")
pass_table
```

```python
imd_table = st_merge.pivot_table("Passed", index=["gender", "age_band"], columns="imd_band", margins=True)
imd_table
```

```python
imd_age = st_merge.pivot_table("Passed", index="age_band", columns="imd_band", margins=True)
imd_age
```

```python
sns.relplot(data=imd_age, kind="line")
plt.show()
```
